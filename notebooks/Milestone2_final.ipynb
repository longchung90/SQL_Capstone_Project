## From Raw Data to Strategic Intelligence
**Prepared for: Lobbyists4America**
**Dataset: 1,243,370 congressional tweets, 2008–2017**

## Milestone 1 Recap

We started with two raw files: **1,243,370 tweets** from 548 congressional Twitter accounts, covering 2008–2017. The core challenge was that Twitter stores no political metadata — we knew *who* tweeted but not their party, chamber, or state.

We solved this through a three-step enrichment join against the `@unitedstates` legislators database, recovering party, chamber, and state for **410 of 548 members (74.8%)**. The remaining 138 members (302,898 tweets, 24.4%) are retained in the dataset but excluded from all party and chamber comparisons — a declared, consistent exclusion applied throughout every analysis.

**Final master table:** 1,243,370 rows × 19 columns including `tweet_id`, `user_id`, `text`, `created_at`, `retweet_count`, `favorite_count`, `party`, `chamber`, `state`.

## Section 1 — Descriptive Statistics
### What We Measured and Why

Before testing any hypothesis, we needed to understand the shape and behaviour of the data. The table below summarises every variable examined, why it was chosen, and what the numbers showed.

| Variable | Why We Measured It | Key Finding |
|---|---|---|
| **Retweet count** | Primary measure of influence — how far a tweet travelled | Mean 190, median 4. Mean/median ratio **47.5×** confirms power-law distribution — averages are meaningless |
| **Favorite count** | Secondary engagement metric — how many likes a tweet received | Mean/median ratio **100.4×** — even more extreme; early years (2008–2012) artificially deflated |
| **Text length** | Proxy for communication style | Median 134 chars (ceiling 140). Members consistently fill available space; 15.9% at exactly 140 chars |
| **Party distribution** | Baseline composition check before any grouped comparison | Republicans: 519,268 tweets (41.8%); Democrats: 405,249 (32.6%). Republicans post 28% more but with lower median engagement |
| **Chamber distribution** | Needed to test H2 — Senate vs House influence gap | Senators post 15% more tweets per person than House members AND generate higher median retweets |
| **Tweets per year** | Foundation for H3 — election year spike detection | 2013 (+165.7% YoY) is the largest spike outside early adoption — and it is a **non-election year** |
| **Tweets per member** | Prevents raw totals being misread as signals | Median 2,691 tweets per member; range 4–3,258. Only 4 members active all 10 years — rolling population, not stable panel |
| **Monthly cadence** | Reveals legislative calendar rhythm | August consistently lowest (recess); October election years highest. Directly actionable for campaign timing |

> **Why median, not mean?** The 47.5× mean/median ratio for retweet count means the average is dominated by a handful of viral moments. Throughout this report, **median is the headline engagement figure**. Mean is reported for completeness only.

## Finding 1 — WHO TO TARGET: Influence is concentrated in a handful of members

![chart 10 - Scatter Plot Volume vs High Reach Members](chart10_scatter_volume_vs_reach.png)

**The finding:**
Congressional Twitter influence is not evenly distributed. A small
number of members account for a disproportionate share of all
retweets across the 2008–2017 period. The scatter plot (Chart 10)
shows that the majority of 545 members cluster in the bottom-left
quadrant — high volume, low reach — while a tiny group in the
top-right quadrant drives most of the platform's amplification.

**The evidence:**
- Retweet mean/median ratio of 47.5× confirms extreme concentration
- The top members by total retweets — including Chuck Schumer
  (2.5M retweets from 3,246 tweets) — outperform the median member
  by orders of magnitude
- Senate members consistently outperform House members on median
  retweets per tweet — the chamber advantage is real
- The quadrant framework (Chart 10) provides a ready-made priority
  list: high volume + high reach members are the primary targets

**What this means for a lobbyist:**
A contact strategy focused on the top-right quadrant of Chart 10
— perhaps 50–60 members out of 545 — will capture the overwhelming
majority of congressional Twitter's amplification capacity. Every
hour spent engaging with a bottom-left quadrant member instead is
a measurable opportunity cost.

**Hypothesis 2: Strongly supported.**

## Finding 2 — WHAT TO SAY: Party lines predict topic, not chamber

**The finding:**
Republicans and Democrats use measurably different vocabulary
on Twitter. The bigram analysis (Chart 8) shows clear policy
domain separation — Democrat bigrams cluster around healthcare,
gun violence, climate change and working families; Republican
bigrams cluster around Obamacare, small business, law enforcement
and mobile office hours. The chamber word clouds (Chart 6) showed
that House and Senate members of the same party communicate more
similarly to each other than members of different parties in the
same chamber.
![chart 8 - Bigram Analysis](chart8_bigrams_by_party.png)
**The evidence:**
- "Health care" is the top bigram for *both* parties — but the
  surrounding vocabulary diverges sharply
- Democrat top single words: bill, women, must, health, congress
- Republican top single words: house, bill, thanks, obamacare, president
- Chamber clouds are more similar to each other than party clouds
  are — party is the stronger predictor of topic focus
![chart 6 - WordCloud for Chamber](chart6_wordcloud_chamber.png)
**What this means for a lobbyist:**
Organise your outreach by **party affiliation**, not by chamber.
A message framed around healthcare access will resonate with
Democrat members in both the House and the Senate. The same
message framed around small business regulatory burden will
land with Republican members in both chambers. Targeting by
chamber alone — treating all House members as one audience —
ignores the stronger predictor.

The bipartisan overlap in the bigram lists is the highest-value
intelligence in this finding. Topics that appear prominently in
both party lists represent the policy terrain where cross-party
coalitions are buildable. "Health care" appearing at the top of
both lists is not a coincidence — it is a signal that healthcare
is the single most viable bipartisan lobbying topic in the dataset.

**Hypothesis 1: Partially supported — trending strongly.**

## Finding 3 — WHERE TO TARGET: 28 states offer bipartisan access,
## but geography is highly concentrated

**The finding:**
The three geographic maps (Charts 11–13) reveal that congressional
Twitter influence is as geographically concentrated as it is
concentrated by member. Five states — California, Texas, New York,
Florida and Illinois — dominate total tweet volume. More
importantly, 28 of 50 states have meaningful representation from
both parties, making them genuine bipartisan access markets where
a single public communication strategy can reach both sides
simultaneously. The remaining 22 are effectively single-party.

**The evidence — Chart 11 (Dominant Party):**
The blue-to-red choropleth shows the national partisan geography
of congressional Twitter clearly. California is the most
Democrat-dominant large state (36D vs 12R members). Texas is the
largest Republican-dominant state (21R vs 6D). The Northeast
corridor is a consistent Democrat blue band. The Mountain West
and Deep South are deep Republican red.

**The evidence — Chart 12 (Competition Score):**
This is the most actionable lobbying map in the project. The
competition score — calculated as `1 − |D_share − R_share|` —
identifies states where both parties have genuinely matched
Twitter activity. Darker states are more contested. States with
a score above 0.8 offer genuine two-party access. States below
0.4 are effectively single-party and require traditional
direct-access lobbying rather than public Twitter campaigns.

**The evidence — Chart 13 (Volume + Engagement Bubble):**
Bubble size shows total tweet volume. Bubble colour shows which
party achieves higher median retweets per tweet in that state.
In several states the minority party achieves disproportionate
engagement despite posting fewer tweets — a hidden influence
advantage that the dominant-party map completely obscures. These
minority-party members in otherwise hostile states are some of
the highest-value contacts for a cross-party lobbying campaign.

**What this means for a lobbyist:**
Three maps, three decisions:
![chart 11 - DOMINANT PARTY BY STATE](map1_dominant_party.png)
- **Chart 11** answers: which party controls this state's
  Twitter conversation? Use this to set realistic expectations
  before targeting any state delegation.
![chart 12 - COMPETITION SCORE BY STATE](map2_competition_score.png)
- **Chart 12** answers: where can I run a bipartisan campaign?
  Concentrate public Twitter pressure on the darkest states —
  both parties have active voices there and can be reached
  simultaneously.
![chart 13 - BUBBLE MAP: VOLUME + ENGAGEMENT ADVANTAGE](map3_volume_engagement.png)
- **Chart 13** answers: who punches above their weight?
  The minority-party members in large-bubble states with
  the opposite-colour engagement advantage are the hidden
  high-leverage contacts — widely followed but often overlooked
  because their party does not dominate their state's raw volume.

**Research Question 1: Strongly addressed.**

## Finding 4 — WHEN TO ACT: Legislative crises matter as much
## as elections — and the platform is becoming more powerful

**The finding:**
The 2013 government shutdown produced the largest year-on-year
communication spike in the dataset outside of early Twitter
adoption — larger than most election years. This was not
predicted. It means that major legislative crises drive
congressional Twitter surges as powerfully as electoral cycles,
and that a lobbyist who only plans around election calendars
is missing the highest-intensity communication windows.

A second timing finding: engagement grew faster than volume
from 2015 onward. The typical tweet in 2016 travelled further
than the typical tweet in 2012, even controlling for the
greater number of tweets being posted. The platform is not
just bigger — it is more powerful per tweet.

**The evidence:**
- 2013 YoY growth: +165.7% — the largest spike in the dataset
  outside 2008–2009 early adoption
- The monthly heatmap (Chart 4) shows August consistently
  quiet (congressional recess) and October of election years
  consistently active
- Chart 2 shows median retweets per tweet rising faster than
  tweet volume from 2015 onward
- Democrats and Republicans both spiked in 2013 — confirming
  a structural legislative-crisis effect, not a partisan one

**What this means for a lobbyist:**
Time your public communication campaigns to coincide with
moments of elevated congressional Twitter activity — not just
election years but major legislative flashpoints. The August
recess is the worst window to launch a Twitter-facing campaign.
October of an election year and the weeks surrounding a major
legislative deadline are the best. The platform's growing
engagement rate means that a well-timed campaign in 2016–2017
would have generated far more amplification than the same
campaign in 2012 — and this trend is likely to have continued
beyond the dataset.

**Hypothesis 3: Partially supported — with important nuance.**

## Section 3 — Hypothesis Assessment

### H1 — Party Predicts Topic Focus
**Verdict: Partially supported — trending strongly toward confirmed.**

Visual evidence (Charts 5, 6) and bigram frequency data (Chart 8) are consistent with the hypothesis. The top 15 word lists share only 40% overlap between parties, with the divergent words clustering clearly around different policy domains. Formal TF-IDF normalisation in Milestone 3 will confirm whether the gap is statistically significant.

**Next step:** Year-by-year TF-IDF divergence scores to test whether the vocabulary gap is widening (increasing polarisation) or stable across the decade.

---

### H2 — Influence is Concentrated in a Small Number of Members
**Verdict: Strongly supported. Confidence: High.**

Multiple independent lines of evidence all point to the same conclusion:

- Retweet mean/median ratio of **47.5×** confirms a power-law distribution
- Top 10% of members hold **62.5%** of all retweets; top 20% hold **79.3%**
- Senate median retweets per tweet: **6.0** vs House: **3.0** — a consistent 2× advantage
- Chart 10 scatter plot shows clear quadrant separation with a tiny high-influence cluster

**Next step:** Add Gini coefficient for precision. Test whether the Senate engagement advantage survives after controlling for follower count (senators may simply have larger audiences by virtue of representing whole states).

---

### H3 — Election Years Produce Measurable Activity Spikes
**Verdict: Partially supported — with important nuance. Confidence: Medium.**

Election years do show elevated volume, but the effect is confounded by baseline Twitter adoption growth across the decade. More critically, **the 2013 non-election year produced a larger spike than any election year in the matched dataset** (+165.7% vs average election year growth of +42.7%).

The hypothesis needs to be restated: it is not election years specifically that drive surges — it is **high-stakes political moments**, of which elections are just one type.

> *[See Chart 1 — Volume Over Time, Chart 3 — YoY Growth Rate, Chart 4 — Monthly Heatmap]*

**Next step:** Detrend the time series using log-linear regression to isolate genuine communication surges from baseline adoption growth. Define and test a broader set of "high-stakes political moments" including shutdowns, Supreme Court nominations, and major legislation.

## Section 5 — Additional Questions for Milestone 3

The exploratory analysis surfaced seven questions beyond the original hypotheses. The first five are data-driven; the last two are methodological improvements needed before Milestone 3 hypothesis tests can be run cleanly.

| Question | Source | What Milestone 3 will do |
|---|---|---|
| **Q1** — Do legislative crises produce larger spikes than elections? | 2013 shutdown finding | Event-based spike comparison using detrended time series |
| **Q2** — Does the party vocabulary gap widen over time? | Charts 5 & 8 | Year-by-year TF-IDF divergence score |
| **Q3** — Is the Senate engagement advantage just audience size? | Boxplots / Chart 10 | Regression controlling for follower count |
| **Q4** — Which topics are genuinely bipartisan vs exclusively partisan? | Chart 8 bigrams | Full vocabulary partisan scoring across all bigrams |
| **Q5** — Do reply patterns differ by party? | `is_reply` flag in `tweet_text` | Reply vs broadcast proportion analysis by party, chamber, year |
| **Q6** *(methodological)* — Detrend H3 time series | Chart 3 limitation | Log-linear residual analysis to isolate electoral effect |
| **Q7** *(methodological)* — Normalise member comparisons by active window | Only 4 members active all 10 years | Per-active-month metrics to surface hidden high-value contacts |

| Question | Answer | Source |

> **Honest caveat:** This dataset covers 2008–2017. The platform has changed substantially — the 280-character limit, follower base growth, and increased political intensity all postdate the archive. The patterns identified here are likely to have persisted or strengthened, but the specific members, states, and timing windows will have evolved. Milestone 3 will test stability across the decade within the dataset to indicate how confidently these findings can be projected forward.
